```python
df = pd.read_csv(r"C:\Users\DELL\Downloads\Netflix_Dataset.csv")
df.head()
```

```python
print("\nMissing values")
print(df.isnull().sum())
```

```python
df["Director"].fillna("unknown", inplace=True)
```

```python
df.display('Director')
```

```python
print("\nMissing values")
print(df.isnull().sum())
```

```python
df['Cast'].fillna("Unknown",inplace=True)
```

```python
df["Country"].fillna("Unknown",inplace=True)
```

```python
df["Release_Date"].fillna("Unknown",inplace=True)
```

```python
df["Rating"].fillna("Unknown",inplace=True)
```

```python
print("\nMissing values")
print(df.isnull().sum())
```

```python
df.drop_duplicates(subset=['Title','Release_Date'], keep='first', inplace=True)
```

```python
df['Title']
```

```python
# Fix encoding issues
df['Title'] = df['Title'].apply(lambda x: x.encode('latin1', 'ignore').decode('utf-8', 'ignore'))

# Remove extra invisible characters and strip spaces
df['Title'] = df['Title'].str.replace(r'[^\x00-\x7F]+', '', regex=True).str.strip()

# Check sample after cleaning
print(df['Title'].head(50))
```

```python
# Content Type Analysis
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count values
print(df['Category'].value_counts())

# Plot
sns.countplot(data=df, x="Category", palette="Set3")
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
# Count Movies vs TV Shows
category_counts = df['Category'].value_counts()

# Display counts
print("Counts:\n", category_counts)

# Calculate percentage
category_percent = (category_counts / len(df)) * 100
print("\nPercentage Split:\n", category_percent)
```

```python
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90, colors = sns.color_palette("Set3", n_colors=len(category_counts)))
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')

```

```python
df['Release_Date'].dtype
```

```python
df['Release_Date'].head()
```

```python
Release_date_per_year = df['Release_Date'].value_counts()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(x=Release_date_per_year.index, y=Release_date_per_year.values, markers='o')
plt.title("Count of Release Per Year")
plt.xlabel("Year")
plt.ylabel("Release Count")
plt.grid(True)
plt.show()
```

```python
df['Release_year'] = df['Release_Date'].dt.year
```

```python
Release_by_type = df.groupby(['Release_year','Category']).size().reset_index(name='count')
```

```python
sns.lineplot(data=Release_by_type, x='Release_year', y='count', hue='Category')
plt.title("Growth of Movies vs TV Shows on Netflix (by Year)")
plt.xlabel("Year")
plt.ylabel("Number of Release")
plt.grid(True)
plt.legend(title="Category")
plt.show()
```

```python
top10_countries = df['Country'].value_counts().head(10)

country_total = (
    df.groupby('Country')
      .size()
      .reset_index(name='count')
      .sort_values(by='count', ascending=False)
      .head(10))
print(country_total)
plt.figure(figsize=(12,6))
sns.barplot(data=country_total, x='Country', y='count', palette="Set3")
plt.title("Top 10 Countries Producing Netflix Content")
plt.xticks(rotation=45)
plt.ylabel("Number of Titles")
plt.show()
```

```python
country_category = df.groupby(['Country','Category']).size().reset_index(name="count")
```

```python
pivot_table = country_category.pivot(index='Country', columns ='Category', values='count')
```

```python
pivot_table['Total'] = pivot_table['Movie'] + pivot_table['TV Show']
top10 = pivot_table.sort_values(by="Total", ascending = False).head(10)
```
